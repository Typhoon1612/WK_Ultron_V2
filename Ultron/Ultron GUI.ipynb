```python
import tkinter as tk
from Ultron import chatbot_response

# Create the GUI window
window = tk.Tk()
window.title("Ultron Airline Chatbot")

# Create the label for the input field
input_label = tk.Label(window, text="Welcom to Ultron Airline Ticket Info Chatbot\nHow may I help you?", font=("Arial", 14))
input_label.pack()

# Create the input field for the user's query
input_field = tk.Entry(window, font=("Arial", 14), width=50)
input_field.pack(pady=10)

# Create the frame to hold the output field and scrollbar
output_frame = tk.Frame(window)
output_frame.pack(pady=10)

# Create the scrollbar for the output field
scrollbar = tk.Scrollbar(output_frame)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

# Create the output field for the chatbot's response
output_field = tk.Text(output_frame, font=("Arial", 14), wrap=tk.WORD, height=10, yscrollcommand=scrollbar.set)
output_field.pack(side=tk.LEFT, fill=tk.BOTH)

# Configure the scrollbar to work with the output field
scrollbar.config(command=output_field.yview)

# Define a function to handle the user's query
def handle_query(event=None):
    # Get the user's query from the input field
    query = input_field.get()
    # Clear the input field
    input_field.delete(0, tk.END)

    # Get the chatbot's response to the user's query
    response = chatbot_response(query)

    # Display the chatbot's response in the output field
    output_field.insert(tk.END, "You: " + query + "\n")
    output_field.insert(tk.END, "Ultron Airline Chatbot: " + response + "\n\n")
    output_field.see(tk.END)  # scroll to the bottom of the output field

# Bind the Enter key to the handle_query function
window.bind('<Return>', handle_query)

# Create the button to submit the user's query
submit_button = tk.Button(window, text="Send", font=("Arial", 14), command=handle_query)
submit_button.pack(pady=10)

# Run the GUI window
window.geometry("500x500")
window.mainloop()
```

```python
import tkinter as tk
from Ultron import chatbot_response

# Create the GUI window
window = tk.Tk()
window.title("Ultron Airline Chatbot")

# Create the label for the input field
input_label = tk.Label(window, text="Welcome to Ultron Airline Ticket Info Chatbot\nHow may I help you?", font=("Arial", 14))
input_label.pack()

# Create the input field for the user's query
input_field = tk.Entry(window, font=("Arial", 14), width=50)
input_field.pack(pady=10)

# Create the output field for the chatbot's response
output_field = tk.Label(window, font=("Arial", 14), wraplength=400, justify="left", fg="blue")
output_field.pack(pady=10)

# Define a function to handle the user's query
def handle_query(event=None):
    # Get the user's query from the input field
    query = input_field.get()

    # Get the chatbot's response to the user's query
    response = chatbot_response(query)

    # Display the chatbot's response in the output field
    output_field.config(text=response)

# Bind the "Return" key to the input field
input_field.bind("<Return>", handle_query)

# Create the button to submit the user's query
submit_button = tk.Button(window, text="Send", font=("Arial", 14), command=handle_query)
submit_button.pack(pady=10)

# Run the GUI window
window.geometry("500x500")
window.mainloop()
```
